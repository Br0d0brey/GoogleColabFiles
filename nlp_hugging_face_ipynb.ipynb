```python
from transformers import pipeline
```

## Определение тональности текста

```python
classifier = pipeline("sentiment-analysis")
```

Определение тональности текста

```python
results = classifier("I hate natural language processing.")
```

```python
results
```

Пайплайн для определения тональности текста на русском языке. Модель [RuBERT for Sentiment Analysis](https://huggingface.co/blanchefort/rubert-base-cased-sentiment).

```python
classifier = pipeline("sentiment-analysis", model="blanchefort/rubert-base-cased-sentiment")
```

```python
results = classifier("Я ненавижу обработку естественного языка")
```

```python
results
```

## Классификация текст – определение токсичности

```python
classifier = pipeline("text-classification", model="SkolkovoInstitute/russian_toxicity_classifier")
```

```python
results = classifier("Мамкин кубер-докер смузихлеб")
```

```python
results
```

## Автоматический перевод

```python
translator = pipeline("translation_ru_to_en", model="Helsinki-NLP/opus-mt-ru-en")
```

```python
results = translator("Меня зовут Андрей Гришаев и я создаю проект по обработке естественного языка")
```

```python
results
```

## Автоматическая генерация текста

## Ответы на вопросы

```python
qa_model = pipeline("question-answering", model="AlexKay/xlm-roberta-large-qa-multilingual-finedtuned-ru")
```

```python
context="""Меня зовут Андрей Гришаев, я занимаюсь разработкой на Python.
           Сейчас я работаю над новым проектом с использованием нейронной сети для обработки натурального языка. Так же я изучаю курсы по программированию. Еще люблю кататься на снуборде и коньках, зима - моё любимое время года.
         """
```

```python
question="Какое любимое время года у Андрея Гришаева?"
```

```python
results = qa_model(question = question, context = context)
```

```python
results
```
